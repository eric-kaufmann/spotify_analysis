# Analysis of Spotify data using pandas

```python
import pandas as pd
import random
```

## 1. Preprocessing
### 1.1 Load and transform Playlist data

```python
# pldf = playlist data frame
pldf = pd.read_csv('./data/playlists.csv', error_bad_lines = False, warn_bad_lines = False)
pldf.head()
```

```python
# edit column names
pldf.columns = pldf.columns.str.replace('"', ' ') # remove ""
pldf.columns = pldf.columns.str.replace('name', ' ') # remove "name"
pldf.columns = pldf.columns.str.replace('_id', ' ') # remove _id from user_id
pldf.columns = pldf.columns.str.replace(' ', '') # remove any spaces
pldf.head()
```

```python
pldf.info()
```

```python
# check for NAs
pldf.isnull().sum()
```

```python
# remove lines with NA-values
pldf.dropna(inplace=True)
pldf.reset_index(drop=True, inplace=True)
pldf
```

```python
# check for duplicates
pldf.duplicated().sum()
```

```python
# remove duplicates
pldf.drop_duplicates(inplace=True)
```

### 1.2 Load and transform Song data 

```python
# sdf = song data frame
sdf = pd.read_csv('./data/tracks_features.csv')
sdf.head()
```

```python
sdf.info()
```

```python
# delete unnecessary columns
sdf.drop(columns=['album_id', 'artist_ids'], inplace=True)
```

```python
# rename some columns
sdf.rename(columns={'artists':'artist', 'name':'track'}, inplace=True)
```

```python
# check NAs
sdf.isnull().sum()
```

```python
# check duplicates
sdf.duplicated().sum()
```

```python
# check if id is unique
sdf['id'].duplicated().sum()
```

```python
# transform artists string into list
sdf['artist'] = sdf['artist'].str.strip('[]').str.replace("'", "").str.replace('"', '').str.split(', ')
```

```python
print(f"Song with most artists has index {sdf['artist'].apply(len).idxmax()} with {sdf['artist'].apply(len).max()} artists")
sdf.iloc[1186088]['artist']
```

```python
# explode artists (new line for every artist)
sdf = sdf.explode('artist').reset_index(drop=True)
sdf
```

### 1.3 Merge the two DataFrames and clean data

```python
# merge playlists with song data
merge_df = pd.merge(pldf, sdf, how='left')
merge_df
```

```python
# check how many tracks from playlist data have no matching entry in song data
merge_df.isnull().sum()
```

```python
# remove all rows where there is no matching track
merge_df = merge_df.dropna()
merge_df
```
